# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#1a Get the total rides for each city type
type_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
type_ride_count
```

```python
#  1. Get the total rides for each city type - the instructions say "by name of city,"" but the code has city type
city_ride_count = pyber_data_df.groupby(["city"]).count()["ride_id"]
city_ride_count.head()
```

```python
# 2. Get the total drivers for each city type
drivers_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
city_fare_count = pyber_data_df.groupby(["city"]).sum()["fare"]
city_fare_count.head()
```

```python
#  4. Get the average fare per ride for each city type. 
city_type_avg_fare = pyber_data_df.groupby(["type"]).mean()["fare"]
city_type_avg_fare
```

```python
#5a Get the sum of the fares for each city type. - column 4
sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_driver_type = sum_fares_by_type /drivers_by_city_type
avg_fare_driver_type
```

```python
#  6. Create a PyBer summary DataFrame. 
Pyber_summary_df = pd.DataFrame(
    {"Total Rides": type_ride_count,
     "Total Drivers": drivers_by_city_type,
     "Total Fares": sum_fares_by_type,
     "Average Fare per Ride": city_type_avg_fare,
     "Average Fare per Driver": avg_fare_driver_type})

Pyber_summary_df 
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
Pyber_summary_df = pd.DataFrame(
    {"Total Rides": type_ride_count,
     "Total Drivers": drivers_by_city_type,
     "Total Fares": sum_fares_by_type,
     "Average Fare per Ride": city_type_avg_fare,
     "Average Fare per Driver": avg_fare_driver_type})

Pyber_summary_df.index.name = None

Pyber_summary_df
```

```python
#  8. Format the columns.
Pyber_summary_df = pd.DataFrame(
    {"Total Rides": type_ride_count.map("{:,}".format),
     "Total Drivers": drivers_by_city_type.map("{:,}".format),
     "Total Fares": sum_fares_by_type.map("${:,.2f}".format),
     "Average Fare per Ride": city_type_avg_fare.map("${:,.2f}".format),
     "Average Fare per Driver": avg_fare_driver_type.map("${:,.2f}".format)})

Pyber_summary_df.index.name = None

Pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() on the pyber_data_df to create a new DataFrame showing the sum of the fares 
# for each date where the indices are the city type and date.
fares_city_type_date_df = pyber_data_df.groupby(['type','date']).sum()[['fare']]
fares_city_type_date_df
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_city_type_date_df= fares_city_type_date_df.reset_index()

#1a Get the total rides for each city type
#type_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
#type_ride_count
```

```python
fares_city_type_date_df
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

fares_city_type_date_df_pivot = fares_city_type_date_df.pivot(index='date', columns='type',values='fare')
fares_city_type_date_df_pivot
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fares_city_type_date_df_pivot_df = fares_city_type_date_df_pivot.loc['2019-01-01':'2019-04-29']
fares_city_type_date_df_pivot_df
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fares_city_type_date_df_pivot_df.index = pd.to_datetime(fares_city_type_date_df_pivot_df.index)
fares_city_type_date_df_pivot_df
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
fares_city_type_date_df_pivot_df.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fares_city_type_week = fares_city_type_date_df_pivot_df.resample("W").sum()
fares_city_type_week
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style


# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

#ax = fares_city_type_week.plot()

plt = fares_city_type_week.plot(figsize=(12,4))

#labels
plt.set_title("Total Fare by City Type")
plt.set_ylabel("Fare ($USD)")
plt.legend().set_bbox_to_anchor((0.6,0.6))


#Save plot to analysis folder
plt.figure.savefig("analysis/PyBer_fare_summary.png")

#plt.show()
```
